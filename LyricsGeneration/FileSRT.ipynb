```python
import os
import pandas as pd
from datetime import datetime, date, timedelta
```

```python
# 匯入歌詞動態時間檔案(.srt)
path = 'srt/Imagine-Dragons-Demons-(Official).srt'
song = path.split('/')
filename = song[1]
print('Filename:',filename)
srt_file = open(path, encoding="utf-8")
srt_string = ''.join(srt_file.readlines())
print(srt_string)
srt_file.close()
```

```python
# 用換行符號把每一行切割開，型態為list
srt_list = srt_string.split('\n')
print(type(srt_list))
srt_list
```

```python
# 刪除第一個元素
# https://blog.csdn.net/deqiangxiaozi/article/details/75808863
# https://blog.csdn.net/u012956540/article/details/50816334
#del srt_list[0]
srt_list = srt_list[1:]
srt_list
```

```python
# 在list中删除所有空元素 https://www.cnblogs.com/aloiswei/p/9083638.html
srt_list2 = [i for i in srt_list if i != '']
srt_list2
```

```python
# 在list中刪除特定間隔的元素
# https://ithelp.ithome.com.tw/articles/10204734
del srt_list2[2:len(srt_list):3]
srt_list2
```

```python
len(srt_list2)
```

```python
# 時間檔案描述預處理
timestep1 = srt_list2[2].split(' --> ')
print(timestep1)

start = timestep1[0]
x1 = start.split(':')
y1 = x1[2].split(',')
start_minute = x1[1]
start_second = y1[0]
start_millisecond = y1[1]
print('start = ', start_minute, start_second, start_millisecond)
start_result = '['+ start_minute + ':' + start_second + '.' + str(start_millisecond).zfill(2) + ']'
print(start_result)

end = timestep1[1]
x2 = end.split(':')
y2 = x2[2].split(',')
end_minute = x2[1]
end_second = y2[0]
end_millisecond = y2[1]
print('end = ', end_minute, end_second, end_millisecond)
end_result = '['+ end_minute + ':' + end_second + '.' + str(end_millisecond).zfill(2) + ']'
print(end_result)
```

```python
srt_df = pd.DataFrame()
start_list = []
end_list = []
lyrics_list = []

for i in range(0,len(srt_list2),2):
    timestep1 = srt_list2[i].split(' --> ')
    print(timestep1)

    start = timestep1[0]
    x1 = start.split(':')
    y1 = x1[2].split(',')
    start_minute = x1[1]
    start_second = y1[0]
    start_millisecond = y1[1]
    print('start = ', start_minute, start_second, start_millisecond)
    start_result = '['+ start_minute + ':' + start_second + '.' + str(start_millisecond).zfill(2) + ']'
    #print(start_result)
    start_list.append(start_result)

    end = timestep1[1]
    x2 = end.split(':')
    y2 = x2[2].split(',')
    end_minute = x2[1]
    end_second = y2[0]
    end_millisecond = y2[1]
    print('end = ', end_minute, end_second, end_millisecond)
    end_result = '['+ end_minute + ':' + end_second + '.' + str(end_millisecond).zfill(2) + ']'
    #print(end_result)
    end_list.append(end_result)
        
    lyrics = srt_list2[i+1]
    lyrics_list.append(lyrics)
    
    print('-'*50)

srt_df['start'] = pd.Series(start_list)
srt_df['end'] = pd.Series(end_list)
srt_df['lyrics'] = pd.Series(lyrics_list)

# 輸出原始.srt檔案時間點的csv
srt_path_01 = 'csv/srt_df_01_' + filename
srt_path_01 = srt_path_01.replace('.srt','.csv')
print('Save in ' + srt_path_01)
srt_df.to_csv(srt_path_01, index=False, encoding='utf_8_sig')
srt_df
```

```python
srt_df2 = pd.DataFrame()
t1_list = []
t2_list = []
t3_list = []
result_list = []

for j in range(1,len(srt_df)):
   
    # 空白片段的長度(也就是第一句的起始時間)
    t1 = srt_df['start'][1]
    t1_list.append(t1)
    t1_minutes = int(t1[1:3])
    t1_seconds = int(t1[4:6])
    t1_milliseconds = int(t1[7:10])
    #print(t1_minutes, t1_seconds, t1_milliseconds)
    
    # 每一句的起始時間
    t2 = srt_df['start'][j]
    t2_list.append(t2)
    t2_minutes = int(t2[1:3])
    t2_seconds = int(t2[4:6])
    t2_milliseconds = int(t2[7:10])
    #print(t2_minutes, t2_seconds, t2_milliseconds)
    
    # 第一個音下拍的時間點
    t3 = '[00:01.677]'
    t3_list.append(t3)
    t3_minutes = int(t3[1:3])
    t3_seconds = int(t3[4:6])
    t3_milliseconds = int(t3[7:10])
    #print(t3_minutes, t3_seconds, t3_milliseconds)
    
    t1 = timedelta(minutes=t1_minutes, seconds=t1_seconds, milliseconds=t1_milliseconds)
    t2 = timedelta(minutes=t2_minutes, seconds=t2_seconds, milliseconds=t2_milliseconds)
    t3 = timedelta(minutes=t3_minutes, seconds=t3_seconds, milliseconds=t3_milliseconds)
    # (每一句的起始時間)-(空白片段的長度)+(第一個音下拍的時間點)
    x = t2 - t1 + t3
    
    # 把每一句新的起始時間點的格式做處理
    a = str(x).split(':')
    minute = a[1] # 分鐘
    second = a[2][:2] # 秒鐘
    millisecond = a[2][3:5] # 毫秒
    #print(minute, second, str(millisecond).zfill(2))
    
    # 毫秒轉換成str型態，然後補0補滿兩位數.zfill(2)
    result = '['+ minute + ':' + second + '.' + str(millisecond).zfill(2) + ']'
    #print(result)
    result_list.append(result)
    
    #print('-'*50)

t1_list.insert(0,"[00:00.000]")  
t2_list.insert(0,"[00:00.000]") 
t3_list.insert(0,"[00:00.000]") 
result_list.insert(0,"[00:00.00]") 

srt_df2['t1'] = pd.Series(t1_list)
srt_df2['t2'] = pd.Series(t2_list)
srt_df2['t3'] = pd.Series(t3_list)
srt_df2['result'] = pd.Series(result_list)

# 輸出計算時間差並加上每一行歌詞新的時間點的csv
srt_path_02 = 'csv/srt_df_02_' + filename
srt_path_02 = srt_path_02.replace('.srt','.csv')
print('Save in ' + srt_path_02)
srt_df2.to_csv(srt_path_02, index=False, encoding='utf_8_sig')
srt_df2
```
