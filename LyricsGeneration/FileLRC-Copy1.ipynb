```python
import os
import pylrc
import pandas as pd
from datetime import datetime, date, timedelta
```

```python
# 匯入歌詞動態時間檔案(.lrc)
path = 'lrc/Bruno-Mars-The-Lazy-Song_simple.lrc'
song = path.split('/')
filename = song[1]
print('Filename:',filename)
lrc_file = open(path, encoding="utf-8")
lrc_string = ''.join(lrc_file.readlines())
print(lrc_string)
lrc_file.close()
```

```python
# 用換行符號把每一行切割開，型態為list
lrc_list = lrc_string.split('\n')
print(type(lrc_list))
lrc_list
```

```python
time_list = []
lyrics_list = []

for l in range(len(lrc_list)):
    time = lrc_list[l][1:9]
    lyrics = lrc_list[l][10:]
    time_list.append(time)
    lyrics_list.append(lyrics)

time_list.remove('[by:Rent')
lyrics_list.remove('nAdviser.com]')
# 移除空值
time_list = [i for i in time_list if i != '']
lyrics_list = [i for i in lyrics_list if i != '']
#print(len(time_list))
#print(len(lyrics_list))

lyrics_list
#time_list
```

```python
lrc_df = pd.DataFrame()
t1_list = []
t2_list = []
t3_list = []
result_list = []

for j in range(0,len(time_list)):
    
    #print('print(time_list[1]) = ', time_list[1])
    t1_list.append(time_list[1])
    timestep1 = time_list[1].split(':')
    t1_minutes = int(timestep1[0])
    t1_seconds = int(timestep1[1][:2])
    t1_milliseconds = int(timestep1[1][3:])*10
    #print(t1_minutes, t1_seconds, t1_milliseconds)

    #print('print(time_list[j]) = ',time_list[j])
    t2_list.append(time_list[j])
    timestep2 = time_list[j].split(':')
    t2_minutes = int(timestep2[0])
    t2_seconds = int(timestep2[1][:2])
    t2_milliseconds = int(timestep2[1][3:])*10
    #print(t2_minutes, t2_seconds, t2_milliseconds)
    
    space = '00:32.57'
    t3_list.append(space)
    timestep3 = space.split(':')
    t3_minutes = int(timestep3[0])
    t3_seconds = int(timestep3[1][:2])
    t3_milliseconds = int(timestep3[1][3:])*10
    #print(t3_minutes, t3_seconds, t3_milliseconds)
    
    t1 = timedelta(minutes=t1_minutes, seconds=t1_seconds, milliseconds=t1_milliseconds)
    t2 = timedelta(minutes=t2_minutes, seconds=t2_seconds, milliseconds=t2_milliseconds)
    t3 = timedelta(minutes=t3_minutes, seconds=t3_seconds, milliseconds=t3_milliseconds)
    x = t2 - t1 + t3
    
    a = str(x).split(':')
    minute = a[1]
    second = a[2][:2]
    millisecond = a[2][3:5]
    #print(minute, second, str(millisecond).zfill(2))
    result = '['+ minute + ':' + second + '.' + str(millisecond).zfill(2) + ']'
    #print(result)
    result_list.append(result)
    
    #print('-'*50)

lrc_df['t1'] = pd.Series(t1_list)
lrc_df['t2'] = pd.Series(t2_list)
lrc_df['t3'] = pd.Series(t3_list)
lrc_df['result'] = pd.Series(result_list)
lrc_df['lyrics'] = pd.Series(lyrics_list)

# 輸出計算時間差並加上每一行歌詞新的時間點的csv
lrc_path_01 = 'csv/lrc_df_01_' + filename
lrc_path_01 = lrc_path_01.replace('.lrc','.csv')
print('Save in ' + lrc_path_01)
lrc_df.to_csv(lrc_path_01, index=False, encoding='utf8')
lrc_df
```

```python
# 只印特定欄位
# 有歌詞
lrc_df2 = lrc_df[['result', 'lyrics']]
# 沒歌詞
#lrc_df2 = lrc_df2[['result']]

lrc_path_02 = 'lrc/output/lrc_' + filename
print('Save in ' + lrc_path_02)

lrc_df3 = pd.DataFrame()
value_list = []
for i in range(len(lrc_df2)):
    # 有歌詞
    value = lrc_df2.iat[i, 0] + lrc_df2.iat[i, 1]
    # 沒歌詞
    #value = lrc_df2.iat[i, 0]
    value_list.append(value)
lrc_df3['content'] = pd.Series(value_list)

# 輸出成lrc:不包括每一列的header
lrc_df3.to_csv(lrc_path_02, index=False, header=None)

# 取代歌詞欄位的「′」符號
text = open(lrc_path_02, "r", encoding="utf-8")
text = ''.join([i for i in text]).replace("′", "'")
edit = open(lrc_path_02,"w",encoding="utf-8")
edit.writelines(text)
edit.close()

lrc_df3
```
