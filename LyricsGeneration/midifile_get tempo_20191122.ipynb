```python
def decrypt(path, key):
    note_dur = open("TXT/parts.txt","w")

    mid  = MidiFile(path)

    msg_lst = []
    note_lst = []
    
    # 印出調號 msg.key
    for i,track in enumerate(mid.tracks):
        for msg in track:
            
            # MIDI檔案所有資訊
            #print(msg)
            
            msg_lst.append(str(msg))
            if msg.type == 'key_signature':
                print('key_signature =',msg.key)
                k = msg.key
    
    # 判斷分析的調號是否和原始輸入相吻合
    if(k != key):
        print("Key doesn't match.")
        exit(0)

    # 取得 notes和 duration
    
    # 取note_off存入txt檔案中
    for al in msg_lst:
        if al.find("note_off") == 0:
            parts = al.split(" ")
            note_dur.write(parts[2] + "," + parts[4] + "\n")
    note_dur.close()

    note_dur = open("TXT/parts.txt","r")
    print('note_dur :'+ '\n' + str(note_dur) + '\n')
    
    # 印出txt檔案中每一筆資料(note和time)
    for notes in note_dur:
        notes = notes.strip('\n')
        print(notes)
        #print('notes[5]',notes[5])
        #print('notes[6]',notes[6])
        note = int(notes[5]+notes[6])
        dur = int(notes[13]+notes[14]+notes[15])
```
