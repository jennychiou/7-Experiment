```
import os
import pandas as pd

%tensorflow_version 1.x

try:
  import gpt_2_simple as gpt2
except:
  !pip3 -q install gpt-2-simple
  import gpt_2_simple as gpt2
```

## Load in text data
ThomasVrancken/lyrics_generation

```
dfs = []
link = ('https://raw.githubusercontent.com/ThomasVrancken/'
        'lyrics_generation/master/songdata_{}.csv')
for i in range(4):
  print(link.format(i))
  dfs.append(pd.read_csv(link.format(i)))

df = pd.concat(dfs).reset_index(drop=True)
```

Create a copy of the files on colaboratory's VM (most convenient way to read them with GPT-2).


```
if not os.path.exists('content'):
    os.makedirs('content')
  
pd.DataFrame({"lyrics": df['text']}).to_csv(os.path.join('content', 'lyrics.csv'), index=False)
```

We then download the pretrained models we will be using.

```
for model_name in ["124M","355M","774M"]:  # Choose from ["124M","355M","774M"]
    gpt2.download_gpt2(model_name=model_name)   # model is saved into current directory under /models/124M/
```

```
learning_rate = 0.0001 # 1e-4
optimizer = 'adam' # adam or sgd
batch_size = 1
model_name = "774M" # has to match one downloaded locally
sess = gpt2.start_tf_sess()
```

```
gpt2.finetune(sess,
              'content/lyrics.csv',
              model_name=model_name,
              sample_every=50,
              save_every=50,
              print_every=10,
              learning_rate=learning_rate,
              batch_size=batch_size,
              restore_from='latest',
              steps=500)   # steps is max number of training steps
```

```
lst_results=gpt2.generate(
    sess,
    prefix="An Old Man Lived in the Village",
    nsamples=10,
    temperature=0.8, # change me
    top_p=0.9, # Change me
    return_as_list=True,
    truncate="<|endoftext|>",
    include_prefix=True
    )
```

```
for res in lst_results:
  print(res)
  print('\n -------//------ \n')
```
