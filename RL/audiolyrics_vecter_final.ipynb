```python
import pandas as pd
import numpy as np
import csv
```

```python
df_a = pd.read_csv('audio_npy_vector.csv')
df_l = pd.read_csv('lyrics_npy_vector.csv')
```

```python
with open ('audiolyrics_vecter_final.csv','w',newline = '') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['track_id','audio_vector','lyrics_vector'])
    
    for i in range(0,11394): #11394
        track_id = df_l['id'][i]
        print(i,'|',track_id)
        
        row2 = df_a[df_a['id'].isin([track_id])]
        
        if len(row2) == 0:
            continue
        
        row_df2 = pd.DataFrame(row2)
        #print(row_df2)
        

        
        audio_vec = row_df2.iat[0,1]
        #print(audio_vec)
        
        lyrics_vec = df_l['npy_vector'][i]
        #print(lyrics_vec)
        
        Table = [[track_id,audio_vec,lyrics_vec]]
        writer.writerows(Table)

print("DONE")
```
