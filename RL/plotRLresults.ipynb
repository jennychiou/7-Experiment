```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_folder = "csv/results/"
filepath = data_folder + user_id + '.csv'
```

```python
user_id = 'chiouchingyi'
filepath = data_folder + user_id + '.csv'
print('路徑：',filepath)
df = pd.read_csv(filepath)
l = []
for i in range(0,len(df)):
    avg = df['avg_score'][i]
    l.append(avg)
x = ['RL1','RL2','RL3','RL4','RL5','RL6','RL7','RL8','RL9','RL10','RL11','RL12','RL13','RL14','RL15','RL16','RL17','RL18','RL19','RL20']
y = [l[0],l[1],l[2],l[3],l[4],l[5],l[6],l[7],l[8],l[9],l[10],l[11],l[12],l[13],l[14],l[15],l[16],l[17],l[18],l[19]]

plt.figure(figsize=(12,5))
plt.plot(x,y,'o-',linewidth=2)
plt.xlabel("RL Round")
plt.ylabel("Average Score")
plt.title('User 1')
plt.ylim(0,5)

plt.grid()
filepath2 = data_folder + user_id + '.jpg'
plt.savefig(filepath2, dpi=200)
plt.show()
```
