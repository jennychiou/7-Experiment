```python
import csv
import pandas as pd
```

```python
#匯入歌曲ID
df1 = pd.read_csv('csv/tracksforRL_trackID.csv')
df1.head()
```

```python
#匯入歌名csv
#csv先用記事本另存成UTF8格式，再加上encoding='utf-8'
#https://blog.csdn.net/lht_okk/article/details/54929103
df2 = pd.read_csv('csv/tracks_features2.csv',encoding='utf-8')
```

```python
#ID重疊者
#https://stackoverflow.com/questions/28901683/pandas-get-rows-which-are-not-in-other-dataframe
common = df1.merge(df2, on=['id'])
common
```

```python
len(common)
```

```python
#匯入歌手csv
df3 = pd.read_csv('csv/tracks_artist.csv',encoding='utf-8')
```

```python
#用歌手ID找對應的歌手name
common['artist_name'] = ''
for i in range(0,len(common)):
    artist_id = common['artist_id'][i]
    row = df3[(df3['id'] == artist_id)]
    name = row.iat[0,1]
    common['artist_name'][i] = name
```

```python
common.head()
```

```python
#匯出統整的檔案
print(len(common))
common.to_csv('csv/trackid+trackname+artistid+artistname.csv',columns=['id','name','artist_id','artist_name'])
```
