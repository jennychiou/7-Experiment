```python
import csv
import pandas as pd
```

```python
# 解决Python:'utf-8' codec錯誤碼
# https://blog.csdn.net/serena9636/article/details/72854039
df1 = pd.read_csv('csv/df_artists(with ID).csv',encoding='ISO-8859-15')
df1
```

```python
df2 = pd.read_csv('csv/df_artists(with ID)2.csv',encoding='ISO-8859-15')
df2
```

```python
# 取交集
df_intersection = pd.merge(df1,df2,on=['Name'])
df_intersection.drop_duplicates(subset=['Name','Id_x','Id_y'], keep='first', inplace=True)
df_intersection.to_csv('csv/df_intersection.csv')
df_intersection
```

```python
# 取差集
df_intersection2 = df_intersection.drop("Id_y", axis = 1)
df_intersection2.rename(columns={'Id_x':'Id'}, inplace = True)
df1 = df1.append(df_intersection2)
df1
# df_difference = df1.drop_duplicates(subset=['Name'],keep='first')
# df_difference.reset_index(drop=True)
# df_difference
```
