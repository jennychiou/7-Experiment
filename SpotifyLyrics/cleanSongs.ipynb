```python
import os
import csv
import time
import pylrc
import shutil
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime, date, timedelta
```

```python
# 刪除歌曲名稱結尾為「_0」(為重複歌曲)
def songs_remove(path):
    allFileList2 = os.listdir(path)
    S_dir_artist = []
    S_dir_artist2 = []
    S_dir_title = []
    for file in allFileList2:
        if os.path.isdir(os.path.join(path,file)):
            S_dir_artist.append(file)
            
            artistPath = path + '/' + file
            artistFileList2 = os.listdir(artistPath)
            
            for file2 in artistFileList2:
                pos = file2.find("_0")
                if os.path.isfile(artistPath + '/' + file2):
                    S_dir_title.append(file2)
                    if pos != -1: # 若有包含「_0」字串的歌曲，則pos值不等於-1；反之，如果沒有包含，則pos值等於1
                        print(artistPath + '/' + file2)
                        os.remove(artistPath + '/' + file2) # 刪除檔案
```

```python
path = r'D:/Anaconda3/Scripts/7 實驗進度/SpotifyLyrics/songs'
songs_remove(path)
```

```python
df_songlist_select = pd.read_excel(r'table/df_songlist_select.xlsx')
df_songlist_select
```

```python
# 存在於df_songlist_select的歌曲(代表有下載到歌詞的)，複製歌曲到另一個資料夾(songs_edit)
def copyfile(df_songlist_select):
    oldname_list = [] # 在songs資料夾的原始檔名
    currentname_list = [] # 複製到songs_edit資料夾的檔名
    newname_list = [] # 欲修改的新檔名
    for i in range(len(df_songlist_select)):
        original = df_songlist_select['Original'][i]
        artist = df_songlist_select['Artist'][i]
        title = df_songlist_select['Title'][i]
        
        filename = title + '.mp3'
        base1 = r'D:/TuneFab Spotify Music Converter/songs/'
        oldpath = base1 + original + '/' + filename # 原始音檔位址
        oldname_list.append(oldpath)
        base2 = r'D:/TuneFab Spotify Music Converter/songs_edit/'
        newpath = base2 + filename # 複製目的地
        currentname_list.append(newpath)
        
        newname = base2 + artist + '_' + filename # 複製後的新檔名，透過renamefile()執行
        newname_list.append(newname)
        
        if os.path.exists(oldpath): # 音檔存在
            message = 'exists'
            shutil.copy(oldpath,newpath)
            print(message + ' - ' + newpath)
        else:
            message = 'not exist' # 音檔不存在
            print(message)
    
    df_copyfile = pd.DataFrame(list(zip(oldname_list,currentname_list,newname_list)),columns=['OldName','CurrentName','NewName'])
    return df_copyfile
```

```python
df_copyfile = copyfile(df_songlist_select)
df_copyfile
```

```python
# 資料夾(songs_edit)的歌曲重新命名，依據「歌手_歌名」格式命名
def renamefile(df_copyfile):
    for i in range(len(df_copyfile)):
        old = df_copyfile['OldName'][i]
        current = df_copyfile['CurrentName'][i]
        new = df_copyfile['NewName'][i]
        try:
            os.rename(current,new)
        except:
            pass
```
