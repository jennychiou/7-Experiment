```python
with open ('lyrics_data/lyrics_trackID_new.csv','w',newline = '') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['num','track_id'])
    for i in range(0,11394):
        x = df['lyrics'][i]
        row = df[df['all'].isin([x])]
        row_df = pd.DataFrame(row)
        track_id = row_df.iat[0,2]
        Table = [[int(x),track_id]]
        writer.writerows(Table)
print("DONE")
```

```python
#avg_vec2修改檔名
#https://www.itread01.com/article/1531280936.html
import os
import pandas as pd
path = "lyrics_data/avg_vec2/"
dirs = os.listdir(path)
lyrics_df = pd.read_csv('lyrics_data/lyrics_trackID_new.csv')
print(lyrics_df['track_id'][0])
print(type(dirs))
```

```python
for i in range(0,11394):
    oldname = path + dirs[i]
    track_id = lyrics_df['track_id'][i]
    newname = path + str(track_id) +".npy"
    os.rename(oldname,newname)
    print(oldname)
    print(newname)
    print('----------------------------------------------')
print('DONE')
```
