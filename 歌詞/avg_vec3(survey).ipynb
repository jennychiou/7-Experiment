```python
def main(add):

    df = pd.read_csv('lyrics_data/survey_top30tfidf_v.csv')

    L1 = []
    for num in range(1,301):
        for col in range(0,30):
            #add = 310 
            df_v = df.iat[add,col]
            a = df_v.replace("\n","").replace('[','').replace(']','').replace('    ',' ').replace('   ',' ').replace('  ',' ')
            b = a.split(' ',302)
            if b[0] == '':
                c = eval(b[num])
            else:
                c = eval(b[num-1])
            L1.append(c)
            #print(c)
    #print(L1)

    n = 30  #大列表中幾個數據组成一個小列表
    L2 = [L1[i:i + n] for i in range(0, len(L1), n)]
    #print(L2)
    final_L = []
    for count in range(0,300):
        element = sum(L2[count])/30
        final_L.append(element)
    #print(final_L)

    #第一首的平均向量
    #print(len(final_L))

    data_folder = "lyrics_data/survry20tracks_l/"
    df2 = pd.read_csv('lyrics_data/survey_top30tfidf.csv')
    var = 0 + add
    print(var)
    num1 = df2.iat[var,0]
    print(num1)
    save_path = data_folder + str(num1) + '.npy'
    print(save_path)
    arr_L = np.array(final_L)
    np.save(save_path,arr_L)
```

```python
for value in range(0,20):
    add = value
    main(add)
```
