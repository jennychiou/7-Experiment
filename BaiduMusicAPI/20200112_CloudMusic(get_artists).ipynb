```python
import os
import re
import csv
import json
import urllib
import requests
import numpy as np
import urllib.parse
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def get_artists(url):
    headers={'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
             'Accept-Encoding': 'gzip, deflate',
             'Accept-Language': 'zh-CN,zh;q=0.9',
             'Connection': 'keep-alive',
             'Cookie': '_iuqxldmzr_=32; _ntes_nnid=0e6e1606eb78758c48c3fc823c6c57dd,1527314455632; '
                       '_ntes_nuid=0e6e1606eb78758c48c3fc823c6c57dd; __utmc=94650624; __utmz=94650624.1527314456.1.1.'
                       'utmcsr=(direct)|utmccn=(direct)|utmcmd=(none); WM_TID=blBrSVohtue8%2B6VgDkxOkJ2G0VyAgyOY;'
                       ' JSESSIONID-WYYY=Du06y%5Csx0ddxxx8n6G6Dwk97Dhy2vuMzYDhQY8D%2BmW3vlbshKsMRxS%2BJYEnvCCh%5CKY'
                       'x2hJ5xhmAy8W%5CT%2BKqwjWnTDaOzhlQj19AuJwMttOIh5T%5C05uByqO%2FWM%2F1ZS9sqjslE2AC8YD7h7Tt0Shufi'
                       '2d077U9tlBepCx048eEImRkXDkr%3A1527321477141; __utma=94650624.1687343966.1527314456.1527314456'
                       '.1527319890.2; __utmb=94650624.3.10.1527319890',
             'Host': 'music.163.com',
             'Referer': 'http://music.163.com/',
             'Upgrade-Insecure-Requests': '1',
             'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) '
                           'Chrome/66.0.3359.181 Safari/537.36'}
    
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, 'html5lib')
    
    count = 0    
    for artist in soup.find_all('a', attrs={'class': 'nm nm-icn f-thide s-fc0'}):
        artist_name = artist.string
        artist_id = artist['href'].replace('/artist?id=', '').strip()
        artist_url = 'https://music.163.com/#/artist?id={}'.format(artist_id)
        print(artist_name + '_' + artist_id)
        try:
            writer.writerow((str(artist_id), artist_name, artist_url))
            count += 1
        except Exception as msg:
            print(msg)       
    return count
```

```python
# 歐美男歌手 id=2001
# 歐美女歌手 id=2002
# 歐美組合/樂隊 id=2003
# 歐美男歌手 A開頭 https://music.163.com/#/discover/artist/cat?id=2001&initial=65
# 歐美女歌手 A開頭 https://music.163.com/#/discover/artist/cat?id=2002&initial=65
# 歐美女歌手 B開頭 https://music.163.com/#/discover/artist/cat?id=2002&initial=66

ls1 = [2001, 2002, 2003]    # id的值
ls2 = [65, 66, 67, 68, 69, 70, 71, 72, 73, 74, 75, 76, 77, 78, 79, 80, 81, 82, 83, 84, 85, 86, 87, 88, 89, 90] # initial的值
# ls1 = [2001]
# ls2 = [65]

# 迴圈抓取改成'a'
csvfile = open('CloudMusic_csv/df_artists.csv', 'a', newline='', encoding="utf_8_sig")
writer = csv.writer(csvfile)
writer.writerow(('artist_id', 'artist_name', 'artist_url'))
    
artist_url = []
count_list = []
for i in ls1:    
    for j in ls2:
        url = 'http://music.163.com/discover/artist/cat?id=' + str(i) + '&initial=' + str(j)
        artist_url.append(url)
        count = get_artists(url)
        count_list.append(count)
        print('Count:' + str(count))
        print('-'*70)
```
