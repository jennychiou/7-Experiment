```python
import requests
res = requests.get('https://www.youtube.com/watch?v=MWM-qVkBzus')
print(res.text)
# https://www.youtube.com/watch?v=ZbZSe6N_BXs
# https://www.youtube.com/watch?v=5yAU52qfYuU
```

```python
import re
m = re.search('layer.config = ({.*?});',res.text)
print(m.group(1))
```

```python
import json
jd = json.loads(m.group(1))
print(jd["args"]["adaptive_fmts"])
```

```python
import urllib
from urllib.parse import urlparse
parsed = urlparse(jd["args"]["adaptive_fmts"])
a = urllib.parse.parse_qs(jd["args"]["adaptive_fmts"])
print(a['url'][0])
```

```python
import shutil
res2 = requests.get(a['url'][0], stream = True)
f = open('music01.mp4', 'wb')
shutil.copyfileobj(res2.raw, f)
f.close()
```
