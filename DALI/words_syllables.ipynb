```python
import DALI as dali_code
import os,re,csv,string,syllables,math
import pandas as pd
import numpy as np
```

```python
# import dataset (DALI_v1.0)
dali_data_path = 'C:/Users/jennychiou/Anaconda3/Scripts/7_Experiments/DALI/DALI_v1.0/'
dali_data = dali_code.get_the_DALI_dataset(dali_data_path, skip=[], keep=[])
dali_data
```

```python
# import the info file of DALI_v1.0
# DALI_ID(NO.)/NAME(songname)/YOUTUBE(YT link)/WORKING(YT access or not)
dali_info = dali_code.get_info(dali_data_path + 'info/DALI_DATA_INFO.gz')
lst = []
for i in range(1,len(dali_info)):
    row = dali_info[i]
    lst.append(row.tolist())
df_dali_info = pd.DataFrame(lst,columns=dali_info[0].tolist()) # column info: array(['DALI_ID', 'NAME', 'YOUTUBE', 'WORKING'])
df_dali_info.sort_values(by=['DALI_ID'], inplace=True)
df_dali_info.reset_index(drop=True, inplace=True)
df_dali_info
```

```python
# calculate syllable count of the word
def syllable_count(word):
    try: # determine the value is nan or not
        c = math.isnan(float(word)) # is nan
    except:
        c = False # not nan
    #print(c,'Not NaN!')
    if c == False:
        word = str(word).lower()
        count = 0
        vowels = "aeiouy"
        try:
            if word[0] in vowels:
                count += 1
            for index in range(1, len(word)):
                if word[index] in vowels and word[index - 1] not in vowels:
                    count += 1
            if word.endswith("e"):
                count -= 1
            if count == 0 and word != '-':
                count += 1
            if count == 0 and word == '-':
                count = 0
        except:
            count = 0
    else:
        count = 0
    return count

for m in range(len(df_dali_info)): # len(df_dali_info)
    a_dali_unique_id = df_dali_info['DALI_ID'][m]
    index = str(m).zfill(4)
    filename = index + '_words_' + a_dali_unique_id + '.xlsx'
    #print('filename:',filename)
    data = pd.read_excel('table/timestamp/'+filename)
    df = pd.DataFrame(data, columns= ['text','freq_0','freq_1','time_0','time_1'])

    lst_count = []
    for r in range(len(df)):
        word = df['text'][r]
        count = syllable_count(word)
        lst_count.append(count)
        #print(lst_count)
        #print(len(lst_count))
    df.insert(5, "syllable", lst_count, True) # insert new column('syllable')
    
    try:
        df.to_excel('table/timestamp_syllable/' + str(m).zfill(4) + '_syllables_' + a_dali_unique_id + '.xlsx')
        print(str(m).zfill(4) + ' - Saved - ' + str(m).zfill(4) + '_syllables_' + a_dali_unique_id + '.xlsx')
    except:
        print(str(m).zfill(4)  + ' - Failed - ' + str(m).zfill(4) + '[' + a_dali_unique_id + ']')
```
