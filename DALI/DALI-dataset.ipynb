```python
# https://github.com/gabolsgabs/DALI
import DALI as dali_code
import os,re,csv,string
import pandas as pd
import numpy as np
```

```python
# import dataset (DALI_v1.0)
dali_data_path = 'C:/Users/jennychiou/Anaconda3/Scripts/7_Experiments/DALI/DALI_v1.0/'
dali_data = dali_code.get_the_DALI_dataset(dali_data_path, skip=[], keep=[])
dali_data
```

```python
# import the info file of DALI_v1.0
# DALI_ID(NO.)/NAME(songname)/YOUTUBE(YT link)/WORKING(YT access or not)
dali_info = dali_code.get_info(dali_data_path + 'info/DALI_DATA_INFO.gz')
lst = []
for i in range(1,len(dali_info)):
    row = dali_info[i]
    lst.append(row.tolist())
df_dali_info = pd.DataFrame(lst,columns=dali_info[0].tolist()) # column info: array(['DALI_ID', 'NAME', 'YOUTUBE', 'WORKING'])
df_dali_info.sort_values(by=['DALI_ID'], inplace=True)
df_dali_info.reset_index(drop=True, inplace=True)
df_dali_info
```

```python
# convert all track .gz to a file(df_track_info.csv)
all_track_info = []
for m in range(len(df_dali_info)):
    a_dali_unique_id = df_dali_info['DALI_ID'][m]
    entry = dali_data[a_dali_unique_id]
    info = entry.info
    track_info = []
    id = info['id']
    artist = info['artist']
    title = info['title']
    try:
        genres = info['metadata']['genres']
        genres = ', '.join(genres)
    except:
        genres = ''
    language = info['metadata']['language']
    track_info.append(id)
    track_info.append(artist)
    track_info.append(title)
    track_info.append(genres)
    track_info.append(language)
    all_track_info.append(track_info)
df_track_info = pd.DataFrame(all_track_info,columns=['id','artist','title','genres','language'])
df_track_info.to_excel('table/df_track_info.xlsx')
df_track_info
```

```python
# get words info of each track and output dataframe with words_{track_id}.xlsx
for p in range(len(df_track_info)): # len(df_track_info)
    track_id = df_track_info['id'][p]
    entry = dali_data[track_id] # input each track's id
    annotations = entry.annotations
    all_lst = []
    for j in range (len(annotations['annot']['words'])):
        word = []
        base = annotations['annot']['words'] # use words as unit
        remove_chars = '[!"#$%&\*+,./:;<=>?@，。?★、…【】《》？“”‘’！[\\]^_`{|}~]+' # remove punctuations
        text = base[j]['text']
        text = re.sub(remove_chars, '', text)
        freq_0 = base[j]['freq'][0]
        freq_1 = base[j]['freq'][1]
        time_0 = base[j]['time'][0]
        time_1 = base[j]['time'][1]
        word.append(text)
        word.append(freq_0)
        word.append(freq_1)
        word.append(time_0)
        word.append(time_1)
        all_lst.append(word)
    df_words = pd.DataFrame(all_lst,columns=['text','freq_0','freq_1','time_0','time_1'])
    try:
        df_words.to_excel('table/timestamp/' + str(p).zfill(4) + '_words_' + track_id + '.xlsx')
        print(str(p).zfill(4) + ' - Saved - ' + str(p).zfill(4) + '_words_' + track_id + '.xlsx')
    except:
        print(str(p).zfill(4)  + ' - Failed - ' + str(p).zfill(4) + '[' + track_id + ']')
    #df_words
```

```python
# save line content to lrc_{track_id}.lrc file
def lrc_time(t): # convert second to lrc time format
    seconds = int(t)
    millis = t - seconds
    return '[%02d:%05.02f]' % (seconds / 60, seconds % 60 + millis)

for p in range(len(df_track_info)): # len(df_track_info)
    track_id = df_track_info['id'][p]
    entry = dali_data[track_id] # input each track's id
    annotations = entry.annotations
    all_lst_lines = []
    for j in range (len(annotations['annot']['lines'])):
        line = []
        base = annotations['annot']['lines'] # use words as unit
        remove_chars = '[!"#$%&\*+,./:;<=>?@，。?★、…【】《》？“”‘’！[\\]^_`{|}~]+' # remove punctuations
        text = base[j]['text']
        text = re.sub(remove_chars, '', text)
        text = text.replace(' - ll','ll')
        freq_0 = base[j]['freq'][0]
        freq_1 = base[j]['freq'][1]
        time_0 = base[j]['time'][0]
        time_1 = base[j]['time'][1]
        line.append(text)
        line.append(freq_0)
        line.append(freq_1)
        line.append(time_0)
        line.append(time_1)
        all_lst_lines.append(line)
    
    df_lines = pd.DataFrame(all_lst_lines,columns=['text','freq_0','freq_1','time_0','time_1'])
    #df_lines

    # write content to lrc file
    lst_lrc = []
    for n in range(len(df_lines)):
        lyrics = df_lines['text'][n]
        lyrics = lyrics.replace('- ','').replace('-',' ')
        lyrics += ' \n'
        time = df_lines['time_0'][n]
        time2 = lrc_time(time)
        content = time2 + lyrics
        lst_lrc.append(content)
        #lst_lrc # each track's lyrics contenet
    try:
        lrc_file = open('lrc/' + str(p).zfill(4) + '_lrc_' + track_id + '.lrc','w') 
        lrc_file.writelines(lst_lrc)
        lrc_file.close()
        print(str(p).zfill(4) + ' - Saved - ' + str(p).zfill(4) + '_lrc_' + track_id + '.lrc')
    except:
        print(str(p).zfill(4) + ' - Failed - ' + str(p).zfill(4) + '_lrc_' + track_id + '.lrc')
```
