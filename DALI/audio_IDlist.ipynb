```python
import csv, re, os, math
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, isdir, join
```

```python
# audio(53584218)
mypath = 'audio/'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
print('Count of File:',len(onlyfiles))
onlyfiles
```

```python
# '.mp3'dataframecsv
def clean(x):
    x = x.replace('.mp3','')
    return x
df_audioID = pd.DataFrame(onlyfiles,columns=['filename'])
df_audioID['filename'] = df_audioID['filename'].apply(clean)
df_audioID.to_csv('table/df_audioID.csv',index=0)
df_audioID
```

```python
# IDcsv(df_audioID.csv)
audioID = pd.read_csv('table/df_audioID.csv')
lst_audioID = audioID['filename'].values.tolist()
print(len(lst_audioID))
audioID
```

```python
# xlsx(df_track_info.xlsx)
trackInfo = pd.read_excel('table/df_track_info.xlsx')
trackInfo.drop(['Unnamed: 0'],inplace=True,axis = 1)
trackInfo
```

```python
# trackInfo(5358)IDaudioIDdataframe(df_track_info_select)
# lst_audioIDaudioIDIDlist
df_track_info_select = trackInfo[trackInfo['id'].isin(lst_audioID)].reset_index(drop=True)
df_track_info_select.to_csv('table/df_track_info_select.csv',index=0)
df_track_info_select
```

```python
# NaN
df_track_info_select.isnull().sum(axis=0)
```

```python
df_track_info_select2 = df_track_info_select.dropna().reset_index(drop=True)
df_track_info_select2.to_csv('table/df_track_info_select2.csv',index=0,encoding="utf_8_sig")
df_track_info_select2
```

```python
# 
all_genres = []
for i in range(0,len(df_track_info_select2)): # len(df_track_info_select)
    temp = df_track_info_select2['genres'][i]
    lst_temp = temp.split(", ")
    all_genres.append(lst_temp)
print(len(all_genres))
print(all_genres)
```

```python
# listlist https://www.geeksforgeeks.org/python-convert-a-nested-list-into-a-flat-list/
output = [] 
def removeNestings(l): 
    for i in l: 
        if type(i) == list: 
            removeNestings(i) 
        else: 
            output.append(i)
    return output
output = removeNestings(all_genres)
print(output) # all_genres()
```

```python
# all_genres https://www.geeksforgeeks.org/python-get-unique-values-list/
def unique(list1): 
    unique_list = [] 
    for x in list1: 
        if x not in unique_list: 
            unique_list.append(x)
    return unique_list
    #for x in unique_list: # unique_list
        #print(x)
output_uq = unique(output)
print(len(output_uq))
print(output_uq) # 59
```

```python
#  https://stackoverflow.com/questions/5828123/nested-list-and-count
lst_c = []
for j in range(len(output_uq)):
    lst = output
    c = lst.count(output_uq[j])
    lst_c.append(c)
    #print(output_uq[j] + ' : ' + str(c))

total_dict = dict(zip(output_uq, lst_c))
print(lst_c)
total = sum(lst_c) 
print('Total Count:',total)
```
