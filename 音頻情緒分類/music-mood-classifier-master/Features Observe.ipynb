```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("yourOutput_0309.csv")
df.head(5)
```

```python
#happy
df0 = df[(df.mood==0)]
df0
```

```python
#angry
df1 = df[(df.mood==1)]
df1
```

```python
#sad
df2 = df[(df.mood==2)]
df2
```

```python
#relax
df3 = df[(df.mood==3)]
df3
```

```python
plt.scatter(df0['valence'],df0['energy'],s=5,label='mood0')
plt.scatter(df1['valence'],df1['energy'],s=5,label='mood1')
plt.scatter(df2['valence'],df2['energy'],s=5,label='mood2')
plt.scatter(df3['valence'],df3['energy'],s=5,label='mood3')
plt.legend()
plt.tight_layout()
plt.xlabel('energy')
plt.ylabel('valence')
```
