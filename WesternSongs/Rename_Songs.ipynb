```python
# 查詢檔名包含指定字串(指定字串,根目錄)
# https://www.itread01.com/p/479361.html
def myfind(pattern,root='test/'):
    fix_list = []
    for each in os.listdir(root):
        if os.path.isdir(os.path.join(root,each)):
            myfind(root,os.path.join(root,each))
        if pattern in str(each):
            fix_list.append(os.path.join(root,each))
    print(fix_list)
```

```python
# 刪除lrc不包含「時間點」的動態歌詞檔案
# https://zhidao.baidu.com/question/426891880374997732.html
path="D:/JennyChiou/code TEMP/LrcData_Edit/"
files=os.listdir(path)

def delete_wo_timimg(files):
    count = 0
    for i ,f in enumerate(files):
        with open(path+f, "r",encoding="utf-8") as file:
            data = file.read()
            #print(type(data))
            file.close()
            a = '[' #要查詢的字串
            if a not in data:
                print(f+' not contain')
                count += 1
                os.remove(path+f)
    return count

count = delete_wo_timimg(files)
print('Delete Count:',count)
```

```python
# 輸出歌手、歌名、檔名的csv
path="D:/JennyChiou/code TEMP/LrcData_Edit/"
files=os.listdir(path)

def output_csv(files):
    with open('CSV/SonglistForSearch.csv',"w", newline='',encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Singer','Song','Filename','Search'])
        for i ,f in enumerate(files):
            filename = files[i]
            filename_split = filename.split('_')
            singer = filename_split[0]
            song = filename_split[1].replace('.lrc','')
            search = singer + ' ' + song
            #print(singer+'/'+song+'/'+filename+'/'+search)
            Table = [[singer,song,filename,search]]
            writer.writerows(Table)
output_csv(files)
```
