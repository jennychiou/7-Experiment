```python
import os
import sys
import csv
import time
import string
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

**<font color=blue size=5 face=微軟雅黑>MIDI Music Artist</font>**

```python
# 爬取MIDI Artist
# https://blog.csdn.net/xtingjie/article/details/73465522

base_url = 'https://freemidi.org/artists'
page = requests.get(base_url)
soup = BeautifulSoup(page.text, "html.parser")

# 選擇 class='az-container1'中所有<a>標籤
t1 = soup.find("div", {"class": "az-container1"}).findAll("a")
#print(t1)

df_table = pd.DataFrame()
list_text = []
list_href = []
for t2 in t1:
    t3 = t2.get('href')
    list_text.append(t3[-1])
    list_href.append('https://freemidi.org' + t3)
#print(list_href)

df_table['Type'] = pd.Series(list_text)
df_table['Url'] = pd.Series(list_href)
df_table
```

```python
df_artist = pd.DataFrame()
list_artist = []
list_artist_url = []

for a in range(len(list_href)):
    url = list_href[a]
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    
    # 選擇 class='genre-link-text'中所有<div>標籤
    artist = soup.findAll("div", {"class": "genre-link-text"})
    print(url,':',len(artist))
    
    # 取得<div>標籤中<a>標籤的 href屬性
    for i in range(len(artist)):
        name = artist[i]
        new_name = name.a.get('href')
        list_artist.append(new_name)
        list_artist_url.append('https://freemidi.org/' + new_name)
        
df_artist['ArtistName'] = pd.Series(list_artist)
df_artist['ArtistUrl'] = pd.Series(list_artist_url)
df_artist.to_csv('csv/FreeMidi/df_artist.csv')
df_artist  
```
