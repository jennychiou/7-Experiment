**<font color=blue size=5 face=微軟雅黑>MIDI Music Type Extract</font>**

```python
# 爬取MIDI類別
base_url = 'https://www.8notes.com/piano/'
page = requests.get(base_url)
soup = BeautifulSoup(page.text, "html.parser")

# 選擇 class為 'gs'
results = soup.findAll("a", {"class": "gs"})
#print(results)

list_text = []
list_href = []
df_miditype = pd.DataFrame()

for r in results:
    list_text.append(r.text)
    list_href.append('https://www.8notes.com' + r['href'])
    #print(r.text, r['href'])

df_miditype['Type'] = pd.Series(list_text)
df_miditype['Url'] = pd.Series(list_href)
df_miditype.to_csv(r'csv/df_miditype.csv',index=False)
df_miditype
```

```python
class Get_page():
    def __init__(self):
        # 請求url
        self.url = 'https://www.8notes.com/piano/childrens/?difficulty=2'
    def get_page(self):
        response = requests.get(self.url)
        soup = BeautifulSoup(response.text, "html.parser")
        return soup
    def parse_page(self, soup):
        # 解析網頁內容
        btn = soup.find('div', class_='pagination')
        btn2 = btn.find_all('a', class_='prevnext')
        page_href = btn2[1]['href']
        next_page_url = 'https://www.8notes.com' + page_href
        return next_page_url

if __name__ == '__main__':
    url = 'https://www.8notes.com/piano/childrens/?difficulty=2'
    midi_page = Get_page()
    for page in range(1, 100):
        print('*' * 50)
        print('正在爬取：第 %s 頁' % page)
        content = midi_page.get_page()
        # 循環判斷，content為空表示此頁已經下載完成了,break跳出循環
        if not content == '':
            next_page_url = midi_page.parse_page(soup)
            print(next_page_url)
            sleep = np.random.randint(3,6)
            time.sleep(sleep)
        else:
            print('該類別已下載完最後一頁')
            break
```

```python
base_url = 'https://www.8notes.com/scores/'
i = 13845
path = base_url + str(i) + '.asp?ftype=midi'
print('Path:', path)
page = requests.get(path)
#print(page.text)
soup = BeautifulSoup(page.text, "html.parser")

# 取得下載路徑
results = soup.find("a", {"class": "transp_alphabet"}).attrs["href"]
#print(results)
final_results = 'https://www.8notes.com' + results
print('Download:', final_results)

# 檔名
filename = final_results.split('/')[::-1][0]
print('Filename:', filename)

# 下載 MIDI檔案在資料夾中
mid_file = requests.get(final_results, stream=True)
with open('Midi/' + filename, 'wb') as saveMidFile:
    saveMidFile.write(mid_file.content)
    print('Downloaded {} successfully.'.format(filename))
```
