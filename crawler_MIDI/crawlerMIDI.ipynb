```python
# 爬取MIDI類別
base_url = 'https://www.8notes.com/piano/'
page = requests.get(base_url)
soup = BeautifulSoup(page.text, "html.parser")

# 選擇 class為 'gs'
results = soup.findAll("a", {"class": "gs"})
#print(results)

list_text = []
list_href = []
df_miditype = pd.DataFrame()

for r in results:
    list_text.append(r.text)
    list_href.append('https://www.8notes.com' + r['href'])
    #print(r.text, r['href'])

df_miditype['Type'] = pd.Series(list_text)
df_miditype['Url'] = pd.Series(list_href)
df_miditype.to_csv(r'csv/df_miditype.csv',index=False)
df_miditype
```

```python
# 取出df_miditype欄位的值
df_miditype['Url'][11]
```

```python
page = requests.get(df_miditype['Url'][11])
soup = BeautifulSoup(page.text, "html.parser")

title = soup.findAll("span", {"class": "listbox redtitle"})
title_url = soup.findAll("a", {"class": "listboxrow"})

list_title = []
list_title_url = []
for t,u in zip(title,title_url):
    list_title.append(t.text)
    list_title_url.append(u['href'])

print(list_title)
print(len(list_title))
print(list_title_url)
print(len(list_title_url))
```

```python
# 爬蟲下一頁
# 測試 https://www.8notes.com/piano/childrens/all/?page=1&difficulty=2
url = df_miditype['Url'][11]
print('Current:', url)

# 往下爬 3頁
for i in range(3):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")

    btn = soup.find('div', class_='pagination')
    btn2 = btn.find_all('a', class_='prevnext')

    # 目前頁面網址
    page_href = btn2[1]['href']

    # 下一頁網址
    next_page_url = 'https://www.8notes.com' + page_href
    print('Next', i, ':', next_page_url)
    url = next_page_url
```
