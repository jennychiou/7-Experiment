```python
import pandas as pd
import numpy as np 
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import euclidean_distances
from file_load import file_path,folder_path

path = './dataset'
name = '_OK.npy'
    
file_list, file_name = file_path(path, name)

features_list = []

for npy in file_list:
    feature = np.load(npy)
    features_list.append(feature)

nps = np.array(features_list)

nps2 = cosine_similarity(nps)

np = cosine_distances(nps)
# print(np)
p = pd.DataFrame(np,index = file_name, columns = file_name)
np3 = euclidean_distances(nps)

p
```

```python
data_sheet = pd.DataFrame(nps2,index = file_name,columns=file_name)
data_sheet
```

```python
d = pd.DataFrame(np3,index = file_name,columns=file_name)

for music_name in file_name:
#     print(music_name)
    sort = d[music_name].sort_values(ascending=True)
    sort_20 = pd.DataFrame(sort.head(20))
    print(sort_20)
```
